### 비유사도와 데이터 병합한 파일 + 금리 변동 방향 추가

- 보내드릴 데이터 편집 작업마다 여기에서 별도로 만들어 처리

```python
df = pd.read_csv("./DATA/merged_data.csv", index_col=0)
df
```

```python
df["base_rate_change"] = df["base_rate"] - df["base_rate"].shift(1)  # t에서 t-1을 뺌
df
```

```python
def change_direction(x):
    if x == 0.0:
        return "유지"
    elif x > 0:
        return "상승"
    elif x < 0:
        return "하락"
    else:
        return np.nan

df["base_change_type"] = df["base_rate_change"].apply(change_direction)
df.iloc[:, -2:]
```

```python
df.columns
```

```python
target_cols = ['dissimilarity', 'base_rate', 'base_rate_change', 'base_change_type', 'FFR', 'ko-us_gap', 'governor', 'tr1', 'tr2',]
df = df[target_cols]
df
```
